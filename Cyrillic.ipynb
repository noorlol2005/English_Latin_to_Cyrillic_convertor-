# English Latin to Cyrillic convertor

This Cyrillic Script is inspired from many existing ones:

*   Aa --> Aa
*   Bb --> Бб
*   Cc --> Цц
*   Dd --> Дд
*   Ee --> Ее
*   Ff --> Фф
*   Gg --> Гг
*   Hh --> Хх
*   Ii --> Ии
*   Jj --> Йй
*   Kk --> Кк
*   Ll --> Лл
*   Mm --> Мм
*   Nn --> Нн
*   Oo --> Оо
*   Pp --> Пп
*   Qq --> Ӄӄ (taken from Russian Far Eastern and Siberian scripts)
*   Rr --> Рр
*   Ss --> Сс
*   Tt --> Тт  
*   Uu --> Уу
*   Vv --> Вв
*   Ww --> В̌ в̌ (taken from Cyrillic Wakhi)
*   Xx --> Ққ (taken from Cyrillic script for English made by Niño Eduardo Evan Fernandez inspired by the Serbian alphabet)
*   Yy --> Ꙑꙑ
*   Zz --> Зз

The aim is to help people who use Cyrillic script adjust to the English language's Latin alphabet. An earlier attempt at making a cyrillic script for English was done by Niño Eduardo Evan Fernandez. His alphabet is as follows:

image.png

```python
# Define a dictionary for Latin to Cyrillic mapping
# Add more mappings as needed following the pattern 'Latin_letter': 'Cyrillic_letter'
replacements = {
    'A': 'А', 'a': 'а',
    'B': 'Б', 'b': 'б',
    'C': 'Ц', 'c': 'ц',
    'D': 'Д', 'd': 'д',
    'E': 'E', 'e': 'e',
    'F': 'Ф', 'f': 'ф',
    'G': 'Г', 'g': 'г',
    'H': 'Х', 'h': 'х',
    'I': 'И', 'i': 'и',
    'J': 'Й', 'j': 'й',
    'K': 'К', 'k': 'к',
    'L': 'Л', 'l': 'л',
    'M': 'М', 'm': 'м',
    'N': 'Н', 'n': 'н',
    'O': 'О', 'o': 'о',
    'P': 'П', 'p': 'п',
    'Q': 'Ӄ', 'q': 'ӄ',
    'R': 'Р', 'r': 'р',
    'S': 'С', 's': 'с',
    'T': 'Т', 't': 'т',
    'U': 'У', 'u': 'у',
    'V': 'В', 'v': 'в',
    'W': 'В', 'w': 'в',
    'X': 'Қ', 'қ': 'x',
    'Y': 'Ꙑ', 'y': 'ꙑ',
    'Z': 'З', 'z': 'з',
}

# Take user input
text = input("Enter your text: ")

# Replace characters using the dictionary
for latin_char, cyrillic_char in replacements.items():
    text = text.replace(latin_char, cyrillic_char)

# Print the result
print("Modified text:", text)
```
